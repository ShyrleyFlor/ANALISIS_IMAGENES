Hallar el area del arroz en la imagen



```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Descargar la imagen
!wget https://shorturl.at/2KNIp -O arroz.jpg

# Cargar la imagen
imagen = cv2.imread('arroz.jpg')

# Convertir la imagen a escala de grises
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Aplicar umbralización para obtener una imagen binaria
_, imagen_umbralizada = cv2.threshold(imagen_gris, 128, 255, cv2.THRESH_BINARY_INV)

# Encontrar los contornos
contornos, _ = cv2.findContours(imagen_umbralizada, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Calcular el área total de los contornos
area_total = sum(cv2.contourArea(contorno) for contorno in contornos)

print(f"Área ocupada por el arroz: {area_total:.2f} píxeles cuadrados")

# Mostrar la imagen con los contornos dibujados
imagen_con_contornos = cv2.drawContours(imagen.copy(), contornos, -1, (0, 255, 0), 2)
cv2_imshow(imagen_con_contornos)
```

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Descargar la imagen
!wget https://shorturl.at/CtQKF -O arroz.jpg

# Cargar la imagen
imagen = cv2.imread('arroz.jpg')

# Convertir la imagen a escala de grises
imagen_gris = cv2.cvtColor(imagen, cv2.COLOR_BGR2GRAY)

# Definir la función 'segmenta' utilizando OpenCV para umbralización
def segmenta(X, t):
    # Aplicar umbralización
    _, Y = cv2.threshold(X, t, 255, cv2.THRESH_BINARY)
    # Calcular el área total de los píxeles blancos
    area = np.sum(Y == 255)
    print('Área =', area)
    return Y

# Aplicar la función 'segmenta' con un umbral t=128 (puedes ajustar este valor según sea necesario)
imagen_segmentada = segmenta(imagen_gris, 128)

# Mostrar la imagen segmentada
cv2_imshow(imagen_segmentada)
```
